```python
from ultralytics import YOLO
```

```python
from google.colab import files
uploaded = files.upload()   # choose dataset.zip

!unzip dataset.zip -d /content/
```

```python
yaml_content = """
path: /content/dataset
train: images/train
val: images/val
test: images/test
nc: 1
names: ['eye']
"""

with open("/content/dataset/data.yaml", "w") as f:
    f.write(yaml_content)
```

```python
# Start training from YOLOv8n pretrained weights
model = YOLO("yolov8n.pt")

model.train(
    data="/content/dataset/data.yaml",
    epochs=50,
    imgsz=640
)
```

```python
metrics = model.val(data="/content/dataset/data.yaml")
print(metrics)
```

```python
results = model.predict(source="/content/dataset/images/test", conf=0.25)
for res in results:
    res.show()
```
